```python
# Usecase-2 Display only those customers from CSV_s1, who purchased more than 3 products
```

```python
merged_data= pd.merge(customer, transaction, on='custno', how='inner')
customer_transaction_count = merged_data['custno'].value_counts()
customers_more_than_3_transaction = customer_transaction_count[customer_transaction_count > 3].index
customers_more_than_3_transaction_data = merged_data[merged_data['custno'].isin(customers_more_than_3_transaction)]
print("Customers from CSV_s1 who purchased more than 3 products:")
customers_more_than_3_transaction_data
```

```python
# Usecase-3 Display top 5 most demanded products from CSV_s1:
```

```python
product_demand = transaction['productno'].value_counts().head(5)
get_data=pd.merge(product_demand,product,right_on="productno",left_index=True)
print("Top 5 most demanded products from CSV_s1: ")
get_data
```

```python
# Usecase-4 Display top 5 transactions amount from CSV_s1:
```

```python
trans_amt = transaction["amount"].sort_values(ascending=False).head(5)
trans_amt_data=pd.merge(trans_amt,transaction)
trans_amt_data4=pd.merge(trans_amt_data,customer)
print("Top 5 transactions amount from CSV_s1: ")
trans_amt_data4
```

```python
# Loading PSV (pipe separated values)


paths = 'customers (1).txt'
pipe_customer = pd.read_csv(paths, header=None,sep='|', names = ['custno','firstname','lastname','gender','age','profession','contactNo','emailId','city','state','isActive','createdDate','UpdatedDate'])
pipe_customer.to_csv(paths,index=False)

paths = 'products.txt'
pipe_product = pd.read_csv(paths, header=None,sep='|', names = ['productno', 'productName', 'Description', 'isActive', 'createdDate', 'UpdatedDate'])
pipe_product.to_csv(paths,  index=False)

paths = 'transactions.txt'
pipe_transaction = pd.read_csv(paths , header= None,sep='|', names = ['txnno','txndate','custno','amount','productno','spendby'])
pipe_transaction.to_csv(paths, index=False)
```
